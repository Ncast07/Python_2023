* Es una biblioteca de visualizacion de datos en Python basada en Matplotlib. Y ofrece una interfaz de alto nivel para dibujar graficos.

## Importar la Biblioteca:

```python
#Importar Pandas + Matplotlib + seaborn:

import pandas as pd
import matplotlib as mtp
import matplotlib.pyplot as plt

import seaborn as sns
```

## Comprendiendo la Sintaxis

```python
# Estructura Basica

data = {
    'Data en eje X': [1, 2, 3, 4, 5],
    'Data en eje Y': [2, 4, 6, 8, 10],
    'Variable de Agrupamiento': ['A', 'B', 'A', 'B', 'A']
}

DataFrame = pd.DataFrame(data)

sns.scatterplot(
    data=DataFrame,
    x="Data en eje X",
    y="Data en eje Y",
    hue="Variable de Agrupamiento"
)
```

### 1. Grafico de Linea


```python
print('1. Grafico de Linea:')

x=[40,68,77]
y=[15,20,30]

sns.lineplot(x=x,y=y)
plt.show()
```

### 2. Grafico de Barra

```python
#Funcion Barra

sns.barplot(    x=['A','B','C'],    y=[1,9,4] )

plt.show()
```

```python
#Funcion SET
print("2. Grafico de Barras+ Set")

sns.set(style='darkgrid',palette='dark',font_scale=2)

sns.barplot( x=['A','B','C'], y=[1,9,4] )
plt.show()
```

### 3. Grafico de Dispersion:

```python
print("3. Grafico de Dispersion")

x=[40,68,77]
y=[15,20,30]

sns.scatterplot( x=x,y=y)
plt.show()
```

### Grafico de Histograma

```python
print("4. Grafico de Histograma")

x=[1,3,3,3,3,3,5,7,89,88,88,88,88,88,88,40,40,40,40,40,40,40,40,40,6,1,8,7,1,5,74,1,3,1]

sns.histplot(x=x, bins=10)
plt.show()
```
