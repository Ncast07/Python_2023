## RELATION

```python
#Este tipo de grafico nos va a dar la correlacion entre distintas variables numericas.

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas.head(2)
```

```python
sns.scatterplot(data=propinas, x='total_bill',y='tip')
plt.show()
```

```python
#Segmentamos
sns.scatterplot(data=propinas,
                x='total_bill',
                y='tip',
                hue='day',
                style='time',
                size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
#Segmentamos
markers = {"Lunch":"D","Dinner":"s"}
sns.scatterplot(data=propinas, x='total_bill',y='tip', hue='day', style='time',size='size',markers = markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.relplot(data=propinas, x='total_bill',y='tip', hue='day', style='time',size='size', col='sex')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

# PLOT + PAIRPLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt

propinas = sns.load_dataset("tips")
propinas.head(2)
```

```python
sns.jointplot(
    data=propinas,
    x='total_bill',
    y='tip')
```

```python
sns.jointplot(data=propinas,x='total_bill', y='tip', hue='time')
plt.show()
```

```python
#kind=histograma
sns.jointplot(data=propinas,x='total_bill', y='tip', hue='sex',kind='hist')
plt.show()
```

```python
#marginals
sns.jointplot(data=propinas,x='total_bill', y='tip', hue='sex',kind='hist',
              marginal_ticks=True,
              marginal_kws=dict(bins=25,fill=False,multiple='dodge')) #customizacion
plt.show()
```

```python
#PAIRPLOT
#Sirve para revisar las variables numericas y u relacion entre si.
sns.pairplot(data=propinas)
plt.show()
```

```python
#PAIRPLOT

sns.pairplot(data=propinas, hue='time')
plt.show()
```

# HEATMAP


```python
import seaborn as sns
import matplotlib.pyplot as plt

propinas = sns.load_dataset("tips")
propinas.head(2)
```

```python
#Gracias a Pandas tenemos las correlaciones:

propinas.corr().round(3)
```

```python
#Por suerte seabonr nos da el HEATMAP

sns.heatmap(propinas.corr())
plt.show()
```

```python
#Vamos a darle contexto.
#annot = True, nos da los valores de la correlacion.
#cmap = cambiar los colores. coolwarm/cividis
#linewidths= cambia las lineas entre los cuadrados.
#cbar=False ->quita la barra de correlacion.

sns.heatmap(propinas.corr(),
            annot=True,
            cmap='coolwarm',
            linewidths=5)
plt.show()
```
